```python
import datetime
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import xml.etree.ElementTree as ET
import xmltodict, json
%matplotlib inline
```

##Read in the core parameters taken from Overseer

```python
#ReadData
CoreParams = pd.read_excel('.\CropCoeff.xlsx', sheet_name='CoreParams')
#Fix names
CoreParams['Crop'] = [str(CoreParams.loc[X,'Crop']).replace(" (","_") for X in range(CoreParams.index.size)]
CoreParams['Crop'] = [str(CoreParams.loc[X,'Crop']).replace("(","_") for X in range(CoreParams.index.size)]
CoreParams['Crop'] = [str(CoreParams.loc[X,'Crop']).replace(")","") for X in range(CoreParams.index.size)]
CoreParams['Crop'] = [str(CoreParams.loc[X,'Crop']).replace("/","_") for X in range(CoreParams.index.size)]
CoreParams['Crop'] = [str(CoreParams.loc[X,'Crop']).replace(" ","") for X in range(CoreParams.index.size)]
#SetIndex
CoreParams.set_index('Crop', inplace=True,drop=True)
#some useful lists
SCRUMParamList = ['[Root].MaximumRootDepth.FixedValue',
                '[Root].KLModifier.XYPairs.Y' ,
                '[Stover].Cover.Expanding.SigCoverFunction.Xo.FixedValue' ,
                '[Stover].Cover.Expanding.SigCoverFunction.b.FixedValue',
                '[Stover].Cover.Expanding.SigCoverFunction.Ymax.FixedValue' ,
                '[Stover].XoBiomass.FixedValue',
                '[Stover].bBiomass.FixedValue' ,
                '[Product].ExpectedYield.FixedValue',
                '[Root].RootProportion.FixedValue' ,
                '[Product].HarvestIndex.FixedValue' ,
                '[Product].DryMatterContent.FixedValue' ,
                '[Phenology].CanopyExpanding.Target.FixedValue' ,
                '[Phenology].YieldIncreasing.Target.FixedValue',
                '[Phenology].Senescing.Target.FixedValue',
                '[Stover].MaximumNConc.FixedValue',
                '[Stover].MinimumNConc.FixedValue',
                '[Product].MinimumNConc.FixedValue',
                '[Product].MaximumNConc.FixedValue',
                '[Root].MinimumNConc.FixedValue',
                '[Root].MaximumNConc.FixedValue',
                '[Nodule].FixationRate.FixedValue',
                '[Nodule].FixationMetabolicCost.FixedValue']
CropList = CoreParams.index.tolist()
CropList.sort()
SCRUMParamList.sort()
```

##Set up working dataframes

```python
Step1 = CoreParams.loc[:,~(CoreParams.columns == 'TT_Sen')].copy()
SCRUMParams = Step1.loc[:,~(Step1.columns == 'TT_Mat')].copy()

#SCRUMParams['[Nodule].NFixationOption'] = [str(SCRUMParams.loc[X,'[Nodule].NFixationOption']).replace("'","") for X in range(SCRUMParams.index.size)]
AncillaryParams = pd.DataFrame(index=CoreParams.index)
```

##Do some simple calculations first

```python
#Calculate the inflection point ot the DM sigmoid as 50% of the crops duration
HalfMat = CoreParams.TT_Mat*0.5
SCRUMParams['[Stover].XoBiomass.FixedValue'] = HalfMat.round(decimals=0)
#The curvature parameter of the DM sigmoid is 1/5 of the inflection point
FifthMat = SCRUMParams['[Stover].XoBiomass.FixedValue']*0.2
SCRUMParams['[Stover].bBiomass.FixedValue'] = FifthMat.round(decimals=0)
#The duration of the Phase from the start to end of senescence is the total tt accumulated to mat less that to sen
SCRUMParams['[Phenology].Senescing.Target.FixedValue'] = CoreParams['TT_Mat'] - CoreParams['TT_Sen']                                                       
#Minimum N conc of stover is the same as maximum
SCRUMParams['[Stover].MinimumNConc.FixedValue'] = SCRUMParams['[Stover].MaximumNConc.FixedValue'] 
#Nodule metabolic fixiation cost is zero
SCRUMParams['[Nodule].FixationMetabolicCost.FixedValue'] = 0
```

##Then calculate some additional values that will be needed for other parameters

```python
#The total amount of biomass if the yield of the crop reached the expected value
AncillaryParams['YMax'] = (SCRUMParams['[Product].ExpectedYield.FixedValue']* \
                          SCRUMParams['[Product].DryMatterContent.FixedValue'])/  \
                          (1-SCRUMParams['[Root].RootProportion.FixedValue'])/ \
                          SCRUMParams['[Product].HarvestIndex.FixedValue']

#The amount of DM at maturity will be slightly less than YMax because the sigmoid won't have reached its asymtote by then
AncillaryParams['DMatMatur'] = AncillaryParams['YMax'] * \
                               1/(1+np.exp(-(CoreParams['TT_Mat']-\
                                             SCRUMParams['[Stover].XoBiomass.FixedValue'])/\
                                           SCRUMParams['[Stover].bBiomass.FixedValue']))
#Root DM at this point is a function of root Proportion
AncillaryParams['RootDM'] =  AncillaryParams['DMatMatur'] * SCRUMParams['[Root].RootProportion.FixedValue']
#Product DM at this point is a function of harvest index
AncillaryParams['ProductDM'] = (AncillaryParams['DMatMatur'] - \
                                AncillaryParams['RootDM']) * \
                                SCRUMParams['[Product].HarvestIndex.FixedValue']
#Stover DM is the residual
AncillaryParams['StoverDM'] = AncillaryParams['DMatMatur'] - \
                              AncillaryParams['RootDM'] - \
                              AncillaryParams['ProductDM']
#We are assuming 20% of the stover biomass is produced during the reproductive phase
AncillaryParams['StoverInRepro'] = AncillaryParams['StoverDM'] * 0.2
#So the amount of biomass at the start of the reproductive phase will be 80% of the total stover 
#adjusted to account for root proportion
AncillaryParams['DMatStartRepro'] = (AncillaryParams['StoverDM'] - \
                                    AncillaryParams['StoverInRepro']) * \
                                    (1+SCRUMParams['[Root].RootProportion.FixedValue'])
#Then we can invert the DM sigmoid to work out how much Tt has passed when this amout of biomass has accumulated
AncillaryParams['TtatStartRepro'] = np.log(1/(AncillaryParams['DMatStartRepro']/\
                                          AncillaryParams['DMatMatur'])-1) * \
                                        SCRUMParams['[Stover].bBiomass.FixedValue'] * -1 + \
                                        SCRUMParams['[Stover].XoBiomass.FixedValue']
```

```python
AncillaryParams
```

##Calculate remaining Phenology parameters

```python
#The tt duration of the canopy expanding (vegetative) phase
SCRUMParams['[Phenology].CanopyExpanding.Target.FixedValue'] = AncillaryParams['TtatStartRepro'].round(decimals=0)
#The tt duration of the yield increasing phase (growing reproductive DM and canopy not senescing)
#is what is left of thermal time to start of senescence after the vegetative phase is taken out
tt_yield = CoreParams['TT_Sen'] - SCRUMParams['[Phenology].CanopyExpanding.Target.FixedValue']
SCRUMParams['[Phenology].YieldIncreasing.Target.FixedValue'] = tt_yield.round(decimals=0)
```

##Generate arays of kl modifier values, adjusted for root depth

```python
def StringMaker(DepthInts,RootDepth):
    YArray = ''
    for Depth in DepthInts:
        if Depth > 0:
            YArray += ','
        if Depth < 305:
            klFact = 1
        elif Depth < RootDepth:
            klFact = 1*np.exp(-.002*(1500/(RootDepth-300)*(Depth-300)))
        else:
            klFact = 0
        klFact = np.round(klFact,3)
        YArray += str(klFact)
    return YArray

DepthInts = range(0,1499,50)
SCRUMParams['[Root].KLModifier.XYPairs.Y'] = \
    [StringMaker(DepthInts,SCRUMParams.loc[X,'[Root].MaximumRootDepth.FixedValue']) \
     for X in SCRUMParams.index]
```
